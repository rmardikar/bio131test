_Why do we expect short reads to produce a more fragmented assembly than long reads?_

We expect short reads to produce a more fragmented assembly because each of the reads are small and arbitrary while long reads are much more coherent and therefor produce a less broken up assembly.

_Why does a single-molecule sequencing like PacBio have a higher error rate than Illumina?_

Single-molecule sequencing has a higher error rate than Illumina because PacBio is less thorough with sequencing since it deals with longer reads and is inefficient to go through the same data multiple times. Illumina on the other hand usually deals with smaller sequences and creates multiple contigs for a read while PacBio does not. However, for creating scaffolds, having more coverage is preferable to higher accuracy.

#### Research and writeup:
According to the paper by Adam, the specific species, Pseudomonas frederiksbergensis is one bacterial strain that selectively provides the R-enantiomer of the sulfoxide molecule by catalyzing asymmetric oxidation of sulfides. Other species selectively show the S-enantiomer of this molecule. This is important because two nearly identical organisms showed a propensity for opposing enantioselectivities, an interesting find considering the tested organisms were not genetically modified. Pseudomonas frederiksbergensis is a gram negative strain discovered in a coal-gasification site in Denmark.

The study done by Andersen, et al., compared the 16S sequence of other species in the Pseudomonas genus to that of Pseudomonas frederiksbergensis showed a close relationship between the different species, however the isolates belonged to a different species – hence the new species under the same genus.

Many species in the Pseudomonas genus are known pathogens to animals and plants, some even contributing to spoilage of milk – an issue in the food industry. In animals, certain species (specifically P. aeruginosa) which exports proteins which contribute to its pathogenicity. 

According to the RAST annotation, the sequence of the genome we inputted is most similar to P. fluorescens, and P. syringae pv. phaseolicola. The genome also shows a degradation of isoleucine (the species might be auxotrophic to isoleucine). There are some repeat regions as well that could be interpreted as evidences of horizontal gene transfer in that the genome already had that gene but some regions were repeated in gene transfer. Comparing this genome to that of P. aeruginosa which is a known pathogen reveals that P. frederiksbergensis does not have any antibiotic resistance and any of the virulence genes that P. aeruginosa does which leads us to believe that this species is probably not pathogenic to humans/animals. The annotation also shows that this genome has resistance to some chemicals that are toxic in excess such as cobalt-zinc-cadmium etc., that P. aeruginosa does not have resistance to. As far as the annotation goes, it seems like this genome does not encode for any known antibiotics and toxins – it does however have different tolerances to various toxins that the other strain that we compared to for reference does not have. This does not mean that this species is unlikely to be harmful in general, but it does indicate that it probably does not affect in the same way as the reference species.

CRISPR has been used in Pseudomonas however papers mention using it in species different than this one. It can be inferred that perhaps CRISPR would also be relevant/useful with this particular species as well. On this topic, since cloning does involve selecting for specific colonies based on antibiotic resistances, the species in question will most likely not be resistant to pencillin, chloramphenicol, or tetracycline.
